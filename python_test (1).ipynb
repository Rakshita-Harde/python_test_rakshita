```python
# take a string input and print 'storng' if it contain at least one uppercase , one lower case , 
# one digit otherwise print 'weak'
s = input("Enter a string: ")

has_upper = any(ch.isupper() for ch in s)
has_lower = any(ch.islower() for ch in s)
has_digit = any(ch.isdigit() for ch in s)

if has_upper and has_lower and has_digit:
    print("strong")
else:
    print("weak")
  
```

```python
# write a function that accepts a list and return (total_sum , total_count , average , sum_of_even , sum_of_odd)
def calculate(lst):
    total_sum = 0
    total_count = 0
    sum_even = 0
    sum_odd = 0

    for num in lst:
        total_sum += num
        total_count += 1

        if num % 2 == 0:
            sum_even += num
        else:
            sum_odd += num

    average = total_sum / total_count

    return total_sum, total_count, average, sum_even, sum_odd
    
numbers = [10, 5, 7, 2, 8]
result = calculate(numbers)

print(result) 
```

```python
# write a function that accepts a  list of tuples and returns a list containing the sum pf each tuple,doubling 
#the value for single elemnet tuple
def tuple_sum(lst):
    result = []

    for t in lst:
        if len(t) == 1:        
            result.append(t[0] * 2)
        else:                  
            s = 0
            for x in t:
                s = s + x
            result.append(s)

    return result
data = [(3, 4), (10,), (5, 5, 5), (7,)]
print(tuple_sum(data))
```

```python
# using map and lambda  convert a list so negative number become positive , zero become 'zero' ,
#positive number remain unchanged 
lst = [3, -5, 0, 8, -2, 0]
result = list(map(lambda x: 'zero' if x == 0 else x if x > 0 else x * -1, lst))
print(result)
```

```python
# use filter and lambda to keep only values that are multiple of 3 and also 
# leave a remainder of 1 when divided by 4 

lst = [1, 3, 4, 7, 9, 10, 13, 15, 16, 19, 21]
result = list(filter(lambda x: x % 3 == 0 and x % 4 == 1, lst))
print(result)
```

```python
# use enumarate() to print items where index + value is even 
lst = [3, 5, 2, 8, 7, 4]
for i, v in enumerate(lst):
    if (i + v) % 2 == 0:
        print(i, v)
```

```python
#use reduce to return the difference between the sum of odd and sum of even number 
from functools import reduce
lst = [1, 4, 5, 8, 3, 10]
result = reduce(lambda a, x: a + x if x % 2 == 0 else a - x, lst)
print(result)
```

```python
# convert a tuple into list , remove elements at even position and convert it back to a tuple 

t = (10, 20, 30, 40, 50, 60)
lst = list(t)
new_list = []
for i in range(len(lst)):
    if i % 2 != 0:    
        new_list.append(lst[i])
result = tuple(new_list)
print(result)
```

```python
# given a set , add and remove elemnts, create a new set with each elemnt divided by 2 and print both sets
s = {10, 20, 30, 40}
s.add(50)
s.remove(20)
new_set = set()
for x in s:
    new_set.add(x / 2)
print("Original set:", s)
print("New set:", new_set)
```
